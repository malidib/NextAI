```python
import pandas as pd
import numpy as np
from IPython.display import HTML, display
```

```python
transactions = 'data/transaction_data.csv'
households = 'data/hh_demographic.csv'
coupons_redeemed = 'data/coupon_redempt.csv'
```

```python
df_trans = pd.read_csv(transactions)
df_hh = pd.read_csv(households)
df_coup = pd.read_csv(coupons_redeemed)
```

```python
hh_visits = df_trans.drop_duplicates('BASKET_ID')
```

```python
num_hh_visits = hh_visits[["household_key", "BASKET_ID"]]
num_hh_visits = num_hh_visits.groupby('household_key').count()
num_hh_visits = num_hh_visits.rename(columns={'BASKET_ID':'number_of_visits'})
num_hh_visits
```

```python
new_features = num_hh_visits
```

```python
hh_visits_delta_days = hh_visits[['household_key', 'BASKET_ID', 'DAY']]
hh_visits_delta_days = hh_visits_delta_days.sort_values(by='DAY')
hh_visits_delta_days = hh_visits_delta_days.assign(day_delta = hh_visits_delta_days.groupby('household_key')['DAY'].diff().fillna(0))
```

```python
hh_visits_delta_days = hh_visits_delta_days.drop(['BASKET_ID', 'DAY'], axis=1)
```

```python
hh_visits_avg_delta = hh_visits_delta_days.groupby('household_key').mean()
hh_visits_avg_delta = hh_visits_avg_delta.rename(columns={'day_delta':'avg_visit_delta'})
hh_visits_avg_delta
```

```python
hh_basket_data = df_trans.drop(['PRODUCT_ID', 'DAY', 'TRANS_TIME', 'WEEK_NO'], axis=1)
hh_basket_data = hh_basket_data.groupby(['household_key', 'BASKET_ID'])
hh_basket_data = hh_basket_data.sum()
hh_basket_data
```

```python
hh_basket_data.to_csv('hh_basket_history.csv')
```

```python
hh_avg_basket = hh_basket_data.drop(['STORE_ID', 'RETAIL_DISC', 'COUPON_DISC', 'COUPON_MATCH_DISC'], axis=1)
hh_avg_basket = hh_avg_basket.mean(level='household_key')
hh_avg_basket = hh_avg_basket.rename(columns={'QUANTITY':'avg_quantity', 'SALES_VALUE':'avg_basket_value'})
hh_avg_basket
```

```python
hh_coups = df_coup.drop(['DAY', 'CAMPAIGN'], axis=1)
hh_coups = hh_coups.groupby('household_key')
hh_coups = hh_coups.count()
hh_coups = hh_coups.rename(columns={'COUPON_UPC':'coupons_redeemed'})
hh_coups
```

```python
new_features = pd.concat([num_hh_visits, hh_visits_avg_delta, hh_avg_basket, hh_coups], axis=1)
new_features['coupons_redeemed'].fillna(0, inplace=True)
new_features = np.round(new_features, 2)
new_features
```

```python
new_features.to_csv('hh_behaviour_features.csv')
```
